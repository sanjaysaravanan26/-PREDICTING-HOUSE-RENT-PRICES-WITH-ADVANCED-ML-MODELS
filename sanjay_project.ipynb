```python
replace_dict={
    1910:"chennai",
    2910:"delhi",
    3310:"bangalore",
    1880:"mumbai",
    1700:"hyderabad"
}
data["Area of the house(excluding basement)"]=data["Area of the house(excluding basement)"].replace(replace_dict)
```

```python
 #Define the list of cities
cities = ['chennai', 'delhi', 'bangalore', 'mumbai', 'hyderabad']

# Assign the values by repeating the list to match the DataFrame length
data['Area of the house(excluding basement)'] = cities * (len(data) // len(cities)) + cities[:len(data) % len(cities)]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
data["Area of the house(excluding basement)"]=label_encoder.fit_transform(data["Area of the house(excluding basement)"])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
import joblib
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid={
    "criterion":["mse","friedman_mse","mae"],
    "splitter":["best","random"],
    "max_depth":[None,10,20,30,40,50],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,3]
}
```

```python
print(data.isnull().sum())
print(np.isinf(data).sum())
```

```python
tree_model=DecisionTreeRegressor()
```

```python
grid_tree=GridSearchCV(estimator=tree_model,param_grid=param_grid,cv=5,n_jobs=-1,verbose=2)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
grid_tree.best_params_
```

```python
grid_tree.predict(x_test)
```

```python
tree_preds=grid_tree.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predslr=lr.predict(x_test)
```

```python
mean_absolute_error(y_test,predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel=RandomForestRegressor()
```

```python
param_gridrfr={
    "max_depth":[5,10,15],
    "n_estimators":[2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr=GridSearchCV(rfrmodel,param_gridrfr)
```

```python
gridrfr.fit(x_train,y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions=gridrfr.predict(x_test)
```

```python
mean_absolute_error(y_test,rfrpredictions)
```
